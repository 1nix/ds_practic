## Предсказание уровня удовлетворённости сотрудника

### Описание данных
- `id` — уникальный идентификатор сотрудника;
- `dept` — отдел, в котором работает сотрудник;
- `level` — уровень занимаемой должности;
- `workload` — уровень загруженности сотрудника;
- `employment_years` — длительность работы в компании (в годах);
- `last_year_promo` — показывает, было ли повышение за последний год;
- `last_year_violations` — показывает, нарушал ли сотрудник трудовой договор за последний год;
- `supervisor_evaluation` — оценка качества работы сотрудника, которую дал руководитель;
- `salary` — ежемесячная зарплата сотрудника;
- `job_satisfaction_rate` — уровень удовлетворённости сотрудника работой в компании, целевой признак.

#### Пайплайн

```python
# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE
ohe_pipe = Pipeline([
     ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('simpleImputer_ohe_2', SimpleImputer(missing_values=' ', strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])
ore_pipe = Pipeline([
     ('simpleImputer_ore', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ore', OrdinalEncoder(
                categories=[
                    ['junior','middle','sinior'],
                    ['low','medium','high']
                ],
                handle_unknown='use_encoded_value', 
                unknown_value=np.nan))
])
num_pipe = Pipeline([
     ('simpleImputer', SimpleImputer(missing_values=np.nan, strategy='median')),
     ('num', RobustScaler()),
])

# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer([
     ('ohe', ohe_pipe, ohe_columns),
     ('ore', ore_pipe, ore_columns),
     ('num', num_pipe, num_columns)
    ], 
    remainder='passthrough'
)

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression())
    
])

param_grid = [

    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': [None],
        'models__min_samples_split': [5],
        'models__min_samples_leaf': [5],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    },
    {
        'models':[RandomForestRegressor(random_state=RANDOM_STATE)],
        'models__n_estimators': [100, 200, 300, 400, 450],  # Количество деревьев
        'models__max_depth': [10, 15],  # Максимальная глубина дерева
        'models__min_samples_split': [2, 5, 10,15],  # Минимальное количество образцов для разделения
        'models__min_samples_leaf': [1, 2, 4, 8, 14, 20],  # Минимальное количество образцов в листе
#         'models__max_features': ['auto', 'sqrt', 'log2', 0.5, 0.8],  # Количество признаков для разделения
#         'models__bootstrap': [True, False],  # Использование бутстрапа
#         'models__oob_score': [True, False],  # Использование out-of-bag образцов
#         'models__max_samples': [None, 0.8, 20, 50]  # Количество образцов для обучения каждого дерева

    },
    {
        'models':[Ridge()],
        'models__alpha': range(1,6),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    },

    {
        'models': [LinearRegression(
        )],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  
    }
]


def smape(actual, predicted) -> float: 
  
    if not all([isinstance(actual, np.ndarray),  
                isinstance(predicted, np.ndarray)]): 
        actual = np.array(actual)
        predicted =  np.array(predicted) 
       
  
    return \
        np.mean( 
            np.abs(predicted - actual) / 
            ((np.abs(predicted) + np.abs(actual))/2) 
        )*100
  


score = make_scorer(smape, greater_is_better=False)


# randomized_search = GridSearchCV(
#     pipe_final, 
#     param_grid, 
#     cv=3,
#     scoring=score,
#     n_jobs=-1,
# #     verbose=2
# )
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring=score,
    n_jobs=-1,
    error_score='raise'
)
randomized_search.fit(X_train, y_train)

print ('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', randomized_search.best_score_)

y_test_pred = randomized_search.best_estimator_.predict(X_test)

print(f'Метрика на тестовой выборке SMAPE :{smape(y_test,y_test_pred)}')
print(f'Метрика на тестовой выборке MAE :{mean_absolute_error(y_test,y_test_pred)}')
```

#### Пайплайн

```python
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('simpleImputer_ohe_2', SimpleImputer(missing_values=' ', strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ]
    )
ore_pipe = Pipeline([
    ('simpleImputer_ore', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
    ('simpleImputer_ore_2', SimpleImputer(missing_values=' ', strategy='most_frequent')),
    ('ore', OrdinalEncoder(
                categories=[
                    ['junior','middle','sinior'],
                    ['low','medium','high']
                ],
                handle_unknown='use_encoded_value', 
                unknown_value=np.nan))
    ])
num_pipe = Pipeline(
    [('simpleImputer', SimpleImputer(missing_values=np.nan, strategy='median')),
     ('num', StandardScaler()),
#      ('poly', PolynomialFeatures(degree=2))
    ])


data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ore', ore_pipe, ore_columns),
     ('num', num_pipe, num_columns)
    ], 
    remainder='passthrough'
)

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
#     ('adasyn', ADASYN(random_state=RANDOM_STATE)),
#     ('feature_selection', SelectFromModel(LinearSVC(penalty="l1"))),
#     ('feature_selection', SelectKBest(mutual_info_classif)),
#     ('feature_selection', SelectKBest(f_classif)),
#     ('feature_selection', FeatureUnion([
#         ('select_k_best', SelectKBest(score_func=mutual_info_classif)),  # Отбор признаков с помощью SelectKBest
#         ('feature_selection_2', SelectFromModel(LogisticRegression(random_state=RANDOM_STATE,
#                                                                    solver='saga',
#                                                                    class_weight='balanced',penalty='l2'))),
# #         ('ext_tree', SelectFromModel(ExtraTreesClassifier(n_estimators=50))),
# #         ('rfe', RFE(estimator=GradientBoostingClassifier(random_state=42), n_features_to_select=2)),  # Отбор признаков с помощью RFE
# #         ('pca', PCA(random_state=RANDOM_STATE, n_components="mle", svd_solver="full"))
#     ],n_jobs=-1)),
    ('models', LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced')) 
])



param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE,class_weight='balanced')],
        'models__max_depth': [None],
        'models__max_features': range(2, 10),
        'models__min_samples_split':range(2,6),
        'models__min_samples_leaf':range(1,6),
        'preprocessor__num': [StandardScaler(), RobustScaler(), 'passthrough'],
#         'feature_selection': [PCA(random_state=RANDOM_STATE, n_components="mle", svd_solver="full"),
#                              ]
    },
    {
        'models': [GradientBoostingClassifier(random_state=RANDOM_STATE)],
        'models__n_estimators': [50, 100, 200],
        'models__learning_rate': [0.01, 0.1, 0.5],
        'models__max_depth': [3,5,7],
        'models__min_samples_leaf':range(1,6),
        'models__max_leaf_nodes': [None, 2, 20, 50],
        'preprocessor__num': [StandardScaler(), RobustScaler(), 'passthrough'],
#         'feature_selection': [PCA(random_state=RANDOM_STATE, n_components="mle", svd_solver="full"),

#                              ]
    },
    {
        'models':[RandomForestClassifier(random_state=42)],
        'models__n_estimators': [100, 200, 300],  # Количество деревьев
        'models__max_depth': [5,10, 15, 30],  # Максимальная глубина дерева
        'models__min_samples_split': [2, 5, 10],  # Минимальное количество образцов для разделения
        'models__min_samples_leaf': [1, 2, 4],  # Минимальное количество образцов в листе
#         'models__max_features': ['auto', 'sqrt', 'log2', 0.5, 0.8],  # Количество признаков для разделения
#         'models__bootstrap': [True, False],  # Использование бутстрапа
#         'models__oob_score': [True, False],  # Использование out-of-bag образцов
        'models__max_samples': [None, 0.8, 20, 50],  # Количество образцов для обучения каждого дерева
        'preprocessor__num': [StandardScaler(), RobustScaler(), 'passthrough'],
    }
    ]

# cv_model = GridSearchCV(
#     pipe_final, 
#     param_grid, 
#     cv=3,
#     scoring='roc_auc',
#     n_jobs=-1
# )
cv_model = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=3,
    scoring='roc_auc',
    n_jobs=-1,

)
cv_model.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', cv_model.best_estimator_)
print('Метрика лучшей модели на тренировочной выборке:', cv_model.best_score_)
print()

y_test_proba = cv_model.best_estimator_.predict_proba(X_test)
y_test_pred = cv_model.best_estimator_.predict(X_test)

print(f'Метрика ROC-AUC на тестовой выборке: {roc_auc_score(y_test, y_test_proba[:,1])}')
print(f'Метрика f1_score на тестовой выборке: {f1_score(y_test, y_test_pred)}')

print(f'Матрица ошибок:{sns.heatmap(confusion_matrix(y_test,y_test_pred),annot=True)}')
```
